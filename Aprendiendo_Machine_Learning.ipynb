# Intro to Machine Learning (By Kaggle)


```python
import kagglehub
import pandas as pd
```

```python
#Importando Melbourne Dataset
path = kagglehub.dataset_download("dansbecker/melbourne-housing-snapshot")
path = path + "/melb_data.csv"
melbourne_data = pd.read_csv(path)
```

```python
melbourne_data.describe()
```

### Preguntas
**Esta desviación estandar es poblacional o muestral?**

---

Es muestral, pandas por defecto considera grados de libertad n-1, si se quiere la desviación poblacional se tiene que especificar del siguiente modo:

```python
#Desviación estandar poblacional
stdp =melbourne_data["Price"].std(ddof=0)
print("std poblacional:",stdp)
#Desviación estandar muestral
stdma = melbourne_data["Price"].std(ddof=1)
stdmb = melbourne_data["Price"].std()
if stdma == stdmb:
  print("ambos metodos dan la misma std muestral:", stdma,  stdma == stdmb)
```

**cómo interpreto los cuartiles (quartesrs)**

---
*"Es la cantidad de información correspondiente a una columna que cae en dicho porcentaje"*

Osea, por ejemplo en la tabla de arriba, la columna rooms:


```python
melbourne_data.Rooms.quantile([0, .25, .5, .75, 1])
```

```python
min = melbourne_data.Rooms.min()
max = melbourne_data.Rooms.max()
print(f'A su vez esto nos dice el valor minimo ({min}) y maximo ({max}) para dicha columna.')
```

Podemos interpretar esto como:
- El 25% de las casas tienen 2 habitaciones o menos
- El 50% de las casas tienen 3 habitaciones o menos
- El 75% de las casas tienen 3 habitaciones o menos
- El minimo de habitaciones que tienen las casas es 1 habitaciones (kinda obvious)
- El máximo de habitaciones que tienen las casas es 10.

Otro Ejemplo, con la columna Price:


```python
melbourne_data["Price"].quantile([0, .25, .5, .75, 1]) #Tambien se puede escribir asi
```

Podemos interpretar esto como:
- El 25% de las casas valen $650,000 (o menos).

- El 50% de las casas valen $903,000 (o menos).

- El 75% de las casas valen $1.330,000 (o menos).

- La casa con menor precio del dataset vale: $85,000.

- La casa con mayor precio del dataset vale: $9.000,000.
